```python
MinMaxScaler_features.shape
```

```python
result = []
for coef in range(2,7):
    print("-"*70)
    print(f"n_clusters = {coef}")
    kmeans = KMeans(n_clusters=coef,random_state=10).fit(MinMaxScaler_features)
    print(kmeans.labels_)
    result.append(kmeans.labels_)
```
